```python
import pandas as pd
```

```python
import os
def find_the_way(path,file_format):
    files_add = []
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            if file_format in file:
                files_add.append(os.path.join(r, file))  
    return files_add
```

```python
path="./"
csvs=find_the_way(path,".xlsx")
csvs
```

```python
df=pd.read_excel(csvs[0],usecols=['English sub group name',  'Turkish sub group animals'])
```

```python
troyer_taxonomy={}

for i in df.values:
    temp=i[1]
    temp=temp.lower()
    temp=temp.replace("  "," ")
    temp=temp.replace(" ","")
    temp=temp.replace("\xa0","")
    troyer_taxonomy[i[0]]=temp.split(",")
```

```python
troyer_taxonomy
```

```python
all_list=[]
for i in troyer_taxonomy:
    for ii in troyer_taxonomy[i]:
        if ii not in all_list:
            all_list.append(ii)
```

```python
all_list
```

```python
import csv
```

```python
df=pd.read_csv("animal.csv")
animals=df['PREPROCESSED TEXT'].values
```

```python
subject=df['ID'].values
```

```python
animals
```

```python
animal_dict={}
#assigning a_list
for i in all_list:
    k=[]
    for j in troyer_taxonomy:
        if i in troyer_taxonomy[j]:
            k.append(j)
    animal_dict[i]=k
#reading patient lists of animals
new_animal=[]
for i in animals:
    j=i.replace(" ",",")
    patient=list(j.split(","))
    new_animal.append(patient) 
#if two consecutive words is same remove repetation
#if there is any repetition without consecutive, leave it
animals2=new_animal
animals=[]
for i in new_animal:
    small_list=[]
    for j in range(0,len(i)-1):
        if i[j]==i[j+1]:
            continue
        else:
            small_list.append(i[j])
    small_list.append(i[len(i)-1])       
    animals.append(small_list)

m=0
for i in animals:
    for j in i:
        try:
            for k in animal_dict[j]:
               continue
        except:

            m+=1
            te="unknown"+str(m)
            animal_dict[j]=[te]
#creating Troyer clusters for searching all animals in Troyer animal list and groups
tr_list=[]
flag=0
for i in animals:
    group=[]
    mlist=[]
    flag=0
    ornek=0
    m=0
    for j in i:
        m+=1
        for jj in i[m:len(i)]:
            flag=0
            for k in animal_dict[j]:
                if k in animal_dict[jj]:
                    flag=1
            if flag:
                if group==[]:
                    group=[j,jj]
                    i.pop(0)
     
                else:
                    group+=[jj]
                    flag=0
                    i.pop(0)
            else:
                if group==[]:
                    group+=[j]
                break
        if (j==i[len(i)-1]) and (group==[]):
            group=[j]
        mlist+= [group]
        group=[]

    tr_list.append(mlist)
    
#creating lists for Troyer lexical features
tr_switches=[]
tr_longest_chain=[]
tr_avarage_chain_lenght=[]
tr_singeleton=[]

#creating of Troyer lexical features
for i in range(0,len(tr_list)):
    tr_switches.append((len(tr_list[i])-1))
    tr_longest_chain.append(len((max(tr_list[i], key=len))))
    avarage=0
    single=0
    for ii in tr_list[i] :
        avarage+=len(ii)
        if len(ii)==1:
            single+=1
    tr_avarage_chain_lenght.append(avarage/len(tr_list[i]))
    tr_singeleton.append(single)
            

#writing Troyer_clusters.csv, WordNet_chains.csv and Word2Vec_chains.csv for observing the output
with open("Troyer_clusters.csv", "w", newline="",encoding="utf-8") as f:
    wrt = csv.writer(f)
    wrt.writerow(["subject_number","Troyer_clusters"])
    for i in range(0,len(tr_list)):
        wrt.writerow([subject[i],tr_list[i]])
```

```python
switch_num=[]
mean_cluster=[]
for i in tr_list:
    size=0
    for ii in i:
        size=size+len(ii)
    switch_num.append(len(i)-1)
    mean_cluster.append(size/len(i))
    
    
    
```
